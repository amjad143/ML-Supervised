```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('HousePrice.csv')
df.head()
```

```python
df.info()
```

```python
# Splitting the data
x = df.iloc[0:,0].values.reshape(-1,1)
y = df.iloc[0:,1].values.reshape(-1,1)
```

```python
sns.pairplot(df);
```

```python
sns.heatmap(df.corr(),annot=True);
```

```python
#Plotting the graph
plt.scatter(x,y,color = 'red')
plt.title("Area and Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

# 2. Model Selection

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting the data for training and testing
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
#Visualising the training data
plt.scatter(train_x,train_y,color = 'red')
plt.title("Area and Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
#Visualising the testing data
plt.scatter(test_x,test_y,color = 'red')
plt.title("Area and Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

# Model Selection

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(train_x,train_y)
```

```python
#plottting the data by using regression line
plt.scatter(df['Area'],df['Price'],color ='red')
plt.plot(train_x,model.predict(train_x),color = 'blue')
plt.title("Area and Price")
plt.xlabel("Area")
plt.ylabel("price")
plt.show()
```

# Predicting values

```python
#Predecting Testing value
pred_y_te = model.predict(test_x)
pred_y_te
```

```python
#Predicting train values
pred_y_tr = model.predict(train_x)
pred_y_tr
```

# Predicting R-squared value

```python
from sklearn.metrics import r2_score
```

```python
#r-square for the "Training" data
pred_y = model.predict(train_x)
r_tr = r2_score(train_y,pred_y)
r_tr
```

```python
#r-square for the "Testing" data
pred_y = model.predict(test_x)
r_ts = r2_score(test_y,pred_y)
r_ts
```
